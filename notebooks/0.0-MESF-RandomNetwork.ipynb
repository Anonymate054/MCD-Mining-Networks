```python
import pandas as pd
```

```python
users = {
    'node': [1, 2, 3, 4, 5],
    'label': ['user', 'user', 'user', 'user', 'user'],
    'property.name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'property.age': [30, 35, 28, 32, 27],
    'property.location': ['New York', 'Los Angeles', 'London', 'Paris', 'Berlin']

}

users_df = pd.DataFrame(users)
users_df
```

```python
posts = {
    "node": [6, 7, 8],
    "label": ["post", "post", "post"],
    "property.content": [
        "Great day at the beach!",
        "Exciting news! Launching a new product.",
        "Delicious dinner with friends."
    ],
    "property.timestamp": [
        "2023-05-15 12:00:00",
        "2023-05-14 15:30:00",
        "2023-05-13 20:45:00"
    ]
}

posts_df = pd.DataFrame(posts)
posts_df
```

```python
photos = {
    "node": [9, 10],
    "label": ["photo", "photo"],
    "property.caption": [
        "Sunset at the mountains",
        "City skyline at night"
    ],
    "property.timestamp": [
        "2023-05-15 18:00:00",
        "2022-05-12 21:30:00"
    ]
}

photos_df = pd.DataFrame(photos)
photos_df
```

```python
import random
from itertools import combinations, permutations
```

```python
random.seed(20242508)

friends = []
follows = []
publishes = []
likes = []


# Relación de FRIENDS

# número de combinaciones de usuarios
friends_combinations = list(combinations(users_df['property.name'].tolist(), 2))

number_friends = random.randint(4, len(friends_combinations)) # 4 amistades mínimo

while number_friends > 0:

    pair = random.sample(friends_combinations, 1)

    if pair[0] in friends_combinations:

        friends.append(pair[0])
        friends_combinations.remove(pair[0]) # muestreo sin reemplazo

        number_friends -= 1


# Relación de FOLLOW

# número de permutaciones de usuarios
friends_permutations = list(permutations(users_df['property.name'].tolist(), 2))

number_follows = random.randint(5, len(friends_permutations)) # 5 follows mínimo

while number_follows > 0:

    pair = random.sample(friends_permutations, 1)

    if pair[0] in friends_permutations:

        follows.append(pair[0])
        friends_permutations.remove(pair[0]) # muestreo sin reemplazo

        number_follows -= 1


# Relación de PUBLISHES

for post in posts_df['property.content'].tolist():

    user = random.choice(users_df['property.name'].tolist())
    publishes.append((user, post))

for photo in photos_df['property.caption'].tolist():

    user = random.choice(users_df['property.name'].tolist())
    publishes.append((user, photo))


# Relación de LIKES

likes = {}

posts_list = posts_df['property.content'].tolist()
photos_list = photos_df['property.caption'].tolist()

content = posts_list + photos_list

for user in users_df['property.name'].tolist():

    num_likes = random.randint(1, len(content)) # número de publicaciones likeadas por un usuario

    user_likes = random.sample(content, num_likes)

    likes[user] = user_likes
```
